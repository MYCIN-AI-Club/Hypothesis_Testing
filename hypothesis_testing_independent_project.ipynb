## Hyphotesis: 
H0: There is no deifference between the blues taken and blue cars returned \n
Ha: There is a  difference between blue cars taken and blue cars returned on a weekend for a specific postal code 


```python
data_weekend = data[data['day_type']=='weekend']
data_weekend.describe()
```

```python
#use random sampling to get 
```

```python
data_weekend.groupby('Postal code')
# Stratified sample
strat_data = data.groupby('Postal code', group_keys=False).apply(lambda grouped_subset : grouped_subset.sample(frac=0.01))
strat_data 
```

```python
BlueCars_taken = np.array(list(strat_data['BlueCars_taken_sum']))
BlueCars_returned = np.array(list(strat_data['BlueCars_returned_sum']))
BlueCars_returned
```

```python
Difference = BlueCars_taken - BlueCars_returned
```

```python
from scipy import stats
```

```python
stats.zscore(Difference, )
```

```python
from statsmodels.stats import weightstats as stests
```

```python
ztest ,pval = stests.ztest(Difference)
ztest ,pval
```

```python
if pval>0.05:
    print("P-Value = " +str(pval)+ " Hence We fail to reject the null hypotheis that returning blue cars = to taking")
else:
    print(("P-Value = " +str(pval)+ " Hence we reject the null hypothesis that returning blue cars = to taking"))
```
